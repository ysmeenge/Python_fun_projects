## Battleship (Zeeslag) - simplified version

The computer creates a board that is 4 cells wide and 3 cells high. The rows are labelled 1 to 3 and the columns are labelled A to D. The computer hides a battleship in three random cells in the board. Each battleship occupies exactly one cell. Battleships are not allowed to touch each other horizontally or vertically.

The player aims to find the boats in minimal amount of steps. The player is asked to 'shoot' at cells in the grid by giving the coordinates. If a boat is hit, the computer returns "YOU HIT MY BOAT!" and removes the boat from the board. If the player misses the computer returns "You've MISSED!".

The programs stops if all boats are found and returns the number of shots needed to find all boats.

```python
import random


def display_board( b ):
    print( "  A B C D" )
    for row in range( 3 ):
        print( row+1, end=" ")
        for col in range( 4 ):
            print( b[row][col], end=" " )
        print('\n')

        
def get_coordinates():
    coordinates = input("Where do you want to aim? Give letter and number combination.\n")
    
    while coordinates not in ("A1", "A2", "A3", "B1", "B2", "B3", "C1", "C2", "C3", "D1", "D2", "D3"):
        coordinates = input("This input is incorrect. Please try again.\n")
        
    row = int(coordinates[1]) - 1
    column = ord(coordinates[0]) - ord('A')
    
    return row, column    


def add_boat_to_board(playing_board, coordinates_list, ships):
    
    for i in range(0, ships):
    
        # get coordinates for boat
        boat_coord = random.choice(coordinates_list)
        row = boat_coord[0]
        column = boat_coord[1]

        #print(f"The boat is placed at row  {row} and the column  {column}.")

        # place boat on board and show board
        playing_board[row][column] = 'O'
        #print("This is what the playing board looks like: ")
        #display_board(playing_board)

        # Remove boat coordinates & coordinates that touch from coordinates_list
        coordinates_list.remove(boat_coord)
        if list([row-1, column]) in coordinates_list:
            coordinates_list.remove(list([row-1, column]))
        if list([row+1, column]) in coordinates_list:
            coordinates_list.remove(list([row+1, column]))
        if list([row, column-1]) in coordinates_list:
            coordinates_list.remove(list([row, column-1]))
        if list([row, column+1]) in coordinates_list:
            coordinates_list.remove(list([row, column+1]))



shots_hit = 0
shots_taken = 0
ships = 3
playing_board = [['.', '.', '.', '.'], ['.', '.', '.', '.'], ['.', '.', '.', '.']]    
coordinates_list = [[0,0], [0,1], [0,2], [0,3],[1,0], [1,1], [1,2], [1,3],[2,0], [2,1], [2,2], [2,3] ]


print('This is what the empty board looks like: \n')
display_board(playing_board)
print("When targeting a boat choose a coordinate from this list: A1, A2, A3, B1, B2, B3, C1, C2, C3, D1, D2, D3.")


# Place 3 boats on board
add_boat_to_board(playing_board, coordinates_list, ships)


#print('This is the game board with boats:')
#display_board(playing_board)


while True:
    
    # ask for coordinates
    row, column = get_coordinates()
    
    # Check if a boat was hit
    if playing_board[row][column] == "O":
        shots_hit += 1
        print(f"YOU HIT MY BOAT!\n \n The number of boats left is: {3-shots_hit}. \n")
        playing_board[row][column] = "." # remove boat from board
    else:
        print("You've MISSED!")

    # check if all boats are found yet
    shots_taken += 1 
    if shots_hit == 3:
        print(f"You found all my boats and needed {shots_taken} hits! \n \n END OF GAME!")
        break
```
