# Tic-Tac-Toe

Player X and Player O take turns. Player X starts. The computer ask the row and column the player want to use. If the row and column combination is already used the question is asked again. After each turn the board is shown.
The first player to get 3 in a row (horizontally, vertically or diagonally) wins. If the board is full, and no player got 3 in a row then no one won.

```python

def get_row_col(player, what):
    index = input(f"On which {what} do you want to put your '{player}'? Type 1, 2, or 3. \n")
    while index not in ("1", "2", "3"):
        index = input('ERROR! This input was incorrect. Type 1, 2, or 3.\n')
    index = int(index)-1
    return index


def display_board(board):
    print("This is what the board looks like now: \n")
    for i in range (0,3): 
        print( " ".join([str(x) if x != 0 else x for x in board[i]]))
    print("\n")


def check_win(board, player):
    if (( board[0][0] == player and board[1][1] == player and board[2][2] == player ) or 
        ( board[0][0] == player and board[0][1] == player and board[0][2] == player ) or
        ( board[1][0] == player and board[1][1] == player and board[1][2] == player ) or 
        ( board[2][0] == player and board[2][1] == player and board[2][2] == player ) or 
        ( board[0][0] == player and board[1][0] == player and board[2][0] == player ) or 
        ( board[0][1] == player and board[1][1] == player and board[2][1] == player ) or 
        ( board[0][2] == player and board[1][2] == player and board[2][2] == player ) or 
        ( board[0][2] == player and board[1][1] == player and board[2][0] == player ) ):
        print(f"PLAYER {player} WON!!!")
        return True
    else:
        return False

    
def opponent(player):
    if player == 'X':
        return 'O'
    return 'X'



player = 'X'
board = [['.', '.', '.'], ['.', '.', '.'], ['.', '.', '.']]
move = 0
max_move = 9

while True:
    
    # Ask row and column
    row = get_row_col(player, "row")
    column = get_row_col(player, "column")
    
    # Check if coordinates are availble
    if board[row][column] != ".":
        print("ERROR! This row and column combinations is already in use. Please try again!")
        continue
    
    # Update and display board
    board[row][column] = player
    display_board(board)
    
    # Check for winner
    if check_win(board,player):
        break
    
    # Quit if board is full
    move += 1
    if move == max_move:
        print("The board is full. Nobody won.")
        break    
    
    # Change players
    player = opponent(player)
    print(f"The turn goes to player {player}.")
```
